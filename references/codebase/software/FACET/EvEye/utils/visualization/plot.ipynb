```python
data_path = Path("/mnt/data2T/junyuan/Datasets/FixedCount5000Dataset/train/cached_data")
event_segment = load_event_segment(888, data_path, 5000)
x = np.array(event_segment['x'])
y = np.array(event_segment['y'])
p = np.array(event_segment['p'])
t = np.array(event_segment['t'])
t -= t[0]
```

```python
colors = np.where(p == 0, 'blue', 'red')

trace = go.Scatter3d(
    x=t,
    y=x,
    z=y,
    mode='markers',
    marker=dict(
        size=2,  # Point size
        color=colors,
        opacity=0.8
    )
)

layout = go.Layout(
    scene=dict(
        xaxis=dict(
            showbackground=False,  # Hide background
            showgrid=False,        # Hide grid lines
            zeroline=False,        # Hide zero line
            showline=False,        # Hide axis lines
            showticklabels=False,  # Hide tick labels
            title=''               # Hide title
        ),
        yaxis=dict(
            showbackground=False,
            showgrid=False,
            zeroline=False,
            showline=False,
            showticklabels=False,
            title=''
        ),
        zaxis=dict(
            showbackground=False,
            showgrid=False,
            zeroline=False,
            showline=False,
            showticklabels=False,
            title=''
        )
    )
)

fig = go.Figure(data=[trace], layout=layout)
fig.write_html("/mnt/data2T/junyuan/eye-tracking/images/3d_scatter_plot.html")
# Show the figure
fig.show()
```

```python
from PIL import Image

# Read the image
input_image_path = '/mnt/data2T/junyuan/eye-tracking/images/3D.png'
output_image_path = '/mnt/data2T/junyuan/eye-tracking/images/3d.png'

# Open the image
image = Image.open(input_image_path).convert("RGBA")

# Get image data
data = image.getdata()

# Create a new data list to store modified pixel values
new_data = []

# Replace the white background with transparency
for item in data:
    # Check whether the pixel is white
    if item[:3] == (255, 255, 255):
        # Replace white pixels with transparency
        new_data.append((255, 255, 255, 0))
    else:
        new_data.append(item)

# Update image data
image.putdata(new_data)

# Save the modified image
image.save(output_image_path, "PNG")

print(f"Image saved to {output_image_path}")
```
